```python
def fizz_buzz_error():
    for index in range(1, 101):
        if index % 3 == 0:
            print("FIZZ")
        elif index % 5 == 0:
            print("BUZZ")
        elif index % 3 and index % 5 == 0:
            print("FIZZ BUZZ")
        else:
            print(index)
```

```python
fizz_buzz_error()
```

```python
def fizz_buzz_simple():
    for index in range(1, 101):
        if index % 5 == 0 and index % 3 == 0:
            print("FizzBuzz")
        elif index % 3 == 0:
            print("Fizz")
        elif index % 5 == 0:
            print("Buzz")
        else:
            print(index)
```

```python
fizz_buzz_simple()
```

```python
def fizz_buzz_optimal():
    for index in range(1, 101):
        if index % 15 == 0:
            print("FIZZ BUZZ")
        elif index % 3 == 0:
            print("FIZZ")
        elif index % 5 == 0:
            print("BUZZ")
        else:
            print(index)
```

```python
fizz_buzz_optimal()
```

```python
import timeit
start_simple = timeit.default_timer()
fizz_buzz_simple()
stop_simple = timeit.default_timer()

start_optimal = timeit.default_timer()
fizz_buzz_optimal()
stop_optimal = timeit.default_timer()
```

```python

print('Time-simple: ', stop_simple - start_simple)  
print('Time-optimal: ', stop_optimal - start_optimal)  
```

```python
import timeit
start_simple = timeit.default_timer()
for _ in range(0, 1000):
    fizz_buzz_simple()
stop_simple = timeit.default_timer()

start_optimal = timeit.default_timer()
for _ in range(0, 1000):
    fizz_buzz_optimal()
stop_optimal = timeit.default_timer()

```
